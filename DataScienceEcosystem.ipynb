**Commonly used libraries in Python used by Data Scientists include:**
<br> 1. Pandas
<br> A powerful data manipulation and analysis library that provides data structures like DataFrames for handling structured data. Essential for data cleaning, transformation, and exploration, making it easier to work with large datasets. Certain applications are to filter, group, or aggregate data for analysis.
<br>
<br> Ex:

```python
import pandas as pd

# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)

# Display the DataFrame
print('This is the original DataFrame')
print(df)
print()

# Filter the data
filtered_df = df[df['Age'] > 28]
print('This is the filtered DataFrame')
print(filtered_df)
```

<br> 2. NumPy 
<br> A library for numerical computations in Python, providing support for arrays and matrices. NumPy forms the foundation for many other scientific computing libraries, making numerical calculations efficient and striaghtforward. Certain applications are performing mathematical operations on large datasets using array manipulation.
<br>
<br> Ex:

```python
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Display the original array
print('This is the original array')
print(arr)
print()

# Perform a mathematical operation
squared = arr ** 2

# Display the new array
print('This is the new array')
print(squared)
```

<br> 3. Seaborn
<br> A statistical data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative graphics. It simplifies the creation of complex visualizations and integrates well with Pandas dataframes. Certain applications are creating heatmaps, pair plots, and violin plots for deeper insights into data relationships. 
<br>
<br> Ex:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Dataset
tips = sns.load_dataset("tips")

# Create a boxplot
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title('Total Bill by Day')
plt.show()
```

<br> 4. Matplotlib 
<br> Matplotlib is a plotting library for Python that provides an object-oriented API for embedding plots into applications. It allows for fine-grained control over the appearance of plots and is foundational for other libraries like Seaborn. Certain applications are creating line graphs, bar charts, and histograms to visualize data distributions.
<br>
<br> Ex:

```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y)
plt.title('Sample Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Shows plot, unnecessary in Jupyter Notebook though
plt.show()
```

<br> 1. ApacheSpark
<br> Apache Spark is an open-source distributed computing system designed for processing large datasets across clusters of computers. It offers APIs in several languages, including Python (PySpark), and is built for speed and efficient.
<br> 
<br> Importance:
<br> Big Data Processing: Spark can handle vast amounts of data that exceed the capacity of a single machine, making it ideal for big data applications.
<br> Speed: It processes data in memory, significantly improving the performance of data processing tasks compared to traditional disk-based systems.
<br> Versatility: Spark supports various data processing tasks such as batch processing, stream processing, machine learning, and graph processing.
<br>
<br> Usage:
<br> Data Analysis: Data scientists use Spark for large-scale data analysis, leveraging Spark SQL for querying structured data
<br> Machine Learning: The MLlib library within Spark provides scalable machine learning algorithms for training models on large datasets.
<br> Streaming Data: Spark Streaming allows for real-time data processing, which is useful in scenarios like fraud detection and live analytics.

```python
from pyspark.sql import SparkSession

# Initialize a Spark session
spark = SparkSession.builder.appName("DataAnalysis").getOrCreate()

# Load data into a DataFrame
df = spark.read.csv("path/to/data.csv", header = True, inferSchema = True)

# Perform a simple transformation
result = df.groupBy("column_name").count()

# Show the result
result.show()
```

<br> 2. TensorFlow
<br> TensorFlow is an open-source machine learning library developed by Google. It is widely used for building and training deep learning models, allowing users to implement a range of machine learning algorithms
<br>
<br> Importance:
<br> Deep Learning: TensorFlow is designed specifically for deep learning, making it a popular choice for developing neural networks.
<br> Scalability: It can run on multiple CPUs and GPUs, enabling efficient training of large-scale models.
<br> Ecosysem: TensorFlow has a rich ecosystem that includes tools for model deployment (TensorFlow Serving), mobile (TensorFlow lite), and JavaScript (TensorFlow.js)
<br>
<br> Usage:
<br> Neural Networks: Data scientists utilize TensorFlow to create complex neural network architectures for tasks such as image recognition, natural language processing, and time series forecasting.
<br> Research and Production: TensorFlow is extensively used in both academic research and production environments for various machine learning applications

```python
import tensorflow as tf
form tensorflow import keras

# Load a dataset (e.g., MNIST)
mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocess Data
x_train = x_train.reshape((60000, 28, 28, 1)).astype('float32') / 255
x_test = x_test.reshape((10000, 28, 28, 1)).astype('float32') / 255

# Build a simple model
model = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (28, 28, 1)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation = 'relu'),
    keras.layers.Dense(10, activation = 'softmax')
])

# Compile and train the model
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
model.fit(x_train, y_train, epochs = 5)
```
